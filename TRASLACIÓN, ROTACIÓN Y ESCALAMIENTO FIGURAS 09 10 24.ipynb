**TRASLACIÓN**

```python
plt.plot([1,2,2,1,1],[1,1,2,2,1])
plt.plot([3,4,4,3,3],[3,3,4,4,3])
plt.arrow(2, 2, 1, 1, head_width=0.1, head_length=0.2, fc='r', ec='r')
```

**ROTACIÓN**

```python
#coordenadas de la figura origial
plt.plot([1,2,2,1,1],[1,1,2,2,1])
#grado de rotación
theta = 45  
#funcion para rotar la figura
def rotar(x, y, theta):
    #formulas
    x_rot = x * np.cos(theta) - y * np.sin(theta)
    y_rot = x * np.sin(theta) + y * np.cos(theta)
    return x_rot, y_rot
# Rotar el cuadrado
x_rot, y_rot = rotar(x, y, theta)
#muestreo de las figura rotada
plt.plot(x_rot, y_rot)
```

**ESCALAMIENTO**

```python
#coordenadas de la figura origial 
c=[1,2,2,1,1]
r=[1,1,2,2,1]

#el escalamiento
Sx, Sy = 4, 4

#formula 
cc= [Sx * x for x in c]
rr = [Sy * y for y in r]

#muestra original y escalonada
plt.plot(c,r)
plt.plot(cc,rr)
```
