**ROTACIÓN**

```python
# Coordenadas del triángulo
x = [1, 2, 1.5, 1]
y = [1, 1, 2, 1]

# Grado de rotación
theta= 45 

# Función para rotar la figura
def rotar(x, y, theta):
    # Convertir listas a arrays de numpy
    x = np.array(x)
    y = np.array(y)
    # Aplicar las fórmulas de rotación
    xr = x * np.cos(theta) - y * np.sin(theta)
    yr = x * np.sin(theta) + y * np.cos(theta)
    return xr, yr

# Rotar el triángulo
xx, yy = rotar(x, y, theta)

# Graficar el triángulo original y rotado
plt.plot(x, y)
plt.plot(xx, yy)
```

```python
# Coordenadas de la figura original
n = [1, 2, 2, 1, 1]
y = [1, 1, 2, 2, 1]

# Grado de rotación
theta = 45 
# Función para rotar la figura
def rotar(n, y, theta):
    # Convertir listas a arrays de numpy
    na = np.array(n)
    ya = np.array(y)
    
    # Aplicar las fórmulas de rotación
    nr = na * np.cos(theta) - ya * np.sin(theta)
    yr = na * np.sin(theta) + ya * np.cos(theta)
    
    return nr, yr

# Rotar la figura
xx, yy = rotar(n, y, theta)

# Graficar la figura original y rotada
plt.plot(n, y)
plt.plot(xx, yy)
```

**ESCALAMIENTO**

```python
#coordenadas de la figura origial 
c = [1, 2, 1.5, 1]
r = [1, 1, 2, 1]

#el escalamiento
Sx, Sy = 4, 4

#formula 
cc= [Sx * x for x in c]
rr = [Sy * y for y in r]

#muestra original y escalonada
plt.plot(c,r)
plt.plot(cc,rr)
```
