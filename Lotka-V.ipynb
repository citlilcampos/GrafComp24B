## Ejemplo incial 

### El tiempo se describe mediante dos ecuaciones diferenciales que consideran las 
### tasas de crecimiento y declive de ambas poblaciones y su interacción entre ellas, 
### en donde: 
### a= 0.7 Tasa de crecimiento de la población de presas en ausencia de depredadores 
### b= 0.5 Tasa de depredación, es decir, la proporción de presas que son cazadas por los depredadores.  
### c= 0.8 Tasa de mortalidad de los depredadores en ausencia de presas.  
### e= 0.2 Eficiencia con la que los depredadores convierten presas cazadas en un incremento en su propia población.

```python
%matplotlib inline
import matplotlib.pyplot as plt
from random import *
from numpy import *
import sys
```

```python
a = 0.7; b = 0.5; c = 0.8;  e = 0.2
```

```python
dt = 0.001; max_time = 100
```

```python
#Tiempo inicial y poblaciones
t = 0; x = 1.0; y = 0.5
```

```python
#Listas vacías en las que almacenar el tiempo y las poblaciones
t_list = []; x_list = []; y_list = []
```

```python
# inicializar listas
t_list.append(t); x_list.append(x); y_list.append(y)
```

```python
while t < max_time:
    # calcular nuevos valores para t, x, y
    t = t + dt
    x = x + (a*x - b*x*y)*dt
    y = y + (-c*y + e*x*y)*dt

    # almacenar nuevos valores en listas
    t_list.append(t)
    x_list.append(x)
    y_list.append(y)
```

```python
# Graficar los resultados
p = plt.plot(t_list, x_list, 'r', t_list, y_list, 'g', linewidth = 2)
```
